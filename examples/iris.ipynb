```python
#%matplotlib inline
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target
 
print(iris.keys())
print('\n--- target_names ---')
print(iris['target_names'])
print('\n--- target ---')
print(iris['target'])

print('\n--- feature_names ---')
print(iris['feature_names'])
print('\n--- data ---')
print(iris['data'])

print('\n--- DESCR ---')
print(iris.DESCR)
```

```python

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.4, s=50,
            cmap=plt.cm.Set1, edgecolor='k')
# X[:, 0] = iris.data.T[0]

# feature_names
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')

plt.xlim(x_min, x_max) # 3.8 8.4
plt.ylim(y_min, y_max) # 1.5 4.9
#plt.xticks(())
#plt.yticks(())

plt.plot([4,8], [2.2,4.8], 'go--', color='red', linewidth=2, markersize=5)
#plt.plot([4,8], [2.2,4.8], marker='o', linestyle='dashed', 
#         color='red', linewidth=2, markersize=5)

plt.show()
```

```python

# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))

ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2],
           c=y, s=50, cmap=plt.cm.Set1, edgecolor='k')

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
plt.show()
```
